### Sigmas

```python
def sigma(column):
    sigma = column.std()*3
    sigma_minus = column.mean() - sigma
    sigma_plus = column.mean() + sigma
    
    return (column>= sigma_minus) & (column <= sigma_plus)
```

```python
df_test_sigma=df_test[df_test.apply(lambda x: sigma(x))]
```

### IQR

```python
def IQR_outliner(column):
    Q1_test=column.quantile(0.25)
    Q3_test=column.quantile(0.75)
    IQR_test = IQR_test=Q3_test-Q1_test
    lower_whisker = Q1_test - 1.5*IQR_test
    upper_whisker = Q3_test+1.5*IQR_test
    return (column > lower_whisker) & (column < upper_whisker)
```

```python
df_test_IQR = df_test[df_test.apply(lambda x: IQR_outliner(x))]
```

### Z-score

```python
df_test.apply(stats.zscore)
```

```python
def mod_z(column):
    thresh = 3.5
    MAD = column.mad()
    nominator = column.apply(lambda x: x - column.median())
    z_score = 0.6745*nominator/MAD
    z_score = z_score[(np.abs(z_score)) < thresh]
    return np.abs(z_score)
```

```python
df_z = df_test.apply(lambda x: mod_z(x))
df_test_z = df_test[df_z >= 0]
```
