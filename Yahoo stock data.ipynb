```python
import altair as alt
import pandas as pd
import plotly.graph_objects as go
```

```python
df = pd.read_csv('Yahoo.csv',parse_dates=['Date'])
```

```python
df
```

```python
df['Moving_Average'] = df['Adj Close'].rolling(3).mean()
```

```python
fig = go.Figure(data=[go.Candlestick(x=df['Date'],
                                     open=df['Open'], 
                                     high=df['High'],
                                     low=df['Low'],
                                     close=df['Close'],
                                    name = 'Stock data'), 
                      go.Scatter(x=df['Date'], y=df['Moving_Average'], line=dict(color='blue', width=2),
                                name = 'Moving average')])
fig.update_layout(
    title="Carriage Services (CSV) - stock chart",
    xaxis_title="Date",
    yaxis_title="Market value (USD)",
    legend_title="Legend",
    font=dict(
        family="Courier New, monospace",
        size=18,
        color="RebeccaPurple"
    )
)
fig.show()
```

```python
color_conditions = alt.condition("datum.Open <= datum.Close",
                                 alt.value("green"),
                                 alt.value("red"))
# build chart
chart = alt.Chart(df).encode(x = "Date")

# set title of chart
chart.title = "Carriage Services (CSV) - stock chart"

# set x axis label for chart
chart.encoding.x.title = "Date"

# construct a rule mark using mark_rule() method
rules = chart.mark_rule().encode(
    y = "Low",
    y2 = "High")

# adjust y axis label for rules
rules.encoding.y.title = "Market value (USD)"

# construct bars
bars = chart.mark_bar().encode(
    y="Open",
    y2="Close",
    color = color_conditions)

# construct line
line = alt.Chart(df).mark_line().encode(
    x='Date',
    y='Moving_Average')

# display rules, bars and line charts
(rules + bars + line).properties(height = 300, width = 700)
```
